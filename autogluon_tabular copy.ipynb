```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import datetime as dt

# Load the dataset
data = pd.read_csv("./data/train.csv", )

X = data[['year', 'month', 'day', 'hour', 'min', 'sec', 'lat', 'lon', 'depth', 'class']].values
y = data[['year_as', 'month_as', 'day_as', 'hour_as', 'min_as', 'sec_as', 'lat_as', 'lon_as', 'depth_as', 'class_as']]

data
```

```python
training_df = pd.DataFrame()
training_df[['year', 'month', 'day', 'hour', 'min', 'sec', 'lat', 'lon', 'depth', 'class']] = X
training_df
```

```python
from autogluon.tabular import TabularDataset, TabularPredictor

predictors = []
features_to_predict = ['year_as','month_as', 'day_as', 'hour_as', 'min_as', 'sec_as', 'lat_as', 'lon_as', 'depth_as', 'class_as']

for i in range(len(features_to_predict)):

    training_df[features_to_predict[i]]=y[features_to_predict[i]]
    
    train_data = TabularDataset(training_df)

    predictor = TabularPredictor(label=features_to_predict[i])

    predictor.fit(
        train_data,
        presets="best_quality",
        ag_args_fit={'num_gpus': 1},
        time_limit=400
    )

    predictors.append(predictor)
```

```python
for predictor in predictors:
    print(predictor.model_best)
    print(predictor.fit_summary())
```
